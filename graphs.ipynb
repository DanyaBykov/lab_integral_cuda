```python
import matplotlib.pyplot as plt
import numpy as np

# Data
time_data = {
    'CUDA': [80.6, 91.73333333333333, 67.8],
    'Threads': [16.666666666666668, 294.6666666666667, 36.53333333333333],
    'Thread Safe Queue': [26.2, 3.1, 53.4],
    'Thread Pool': [26.3, 35.5, 64.2]
}

functions = ['Function 1', 'Function 2', 'Function 3']
methods = list(time_data.keys())

x = np.arange(len(functions))
width = 0.2
step = 25

fig, ax = plt.subplots()

for i, (method, times) in enumerate(time_data.items()):
    ax.bar(x + i * width, times, width, label=method)

ax.set_xlabel('Functions')
ax.set_ylabel('Time (ms)')
ax.set_title('Execution Times for Different Methods')
ax.set_xticks(x + width * (len(methods) / 2 - 0.5))
ax.set_xticklabels(functions)
ax.set_yticks(np.arange(0, 325, step))
ax.legend()

plt.show()
```
